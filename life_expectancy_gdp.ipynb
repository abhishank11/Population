```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns 

data = pd.read_csv('all_data.csv')
data.columns = ['country', 'year', 'life_expectancy', 'gdp']
list_country = data['country'].unique()
country_dict = {}

#Creating separate data sets for each country

for country in list_country:
    country_dict[country] = data[data['country'] == country]

#average life expectancy function
def avg_le(data_set):
    le_sub = data_set.life_expectancy
    average = np.mean(le_sub)
    return average

#average GDP function
def avg_gdp(data_set):
    gdp_sub = data_set.gdp
    average = np.mean(gdp_sub)
    return average

#Chile 
chile_data = country_dict['Chile']
china_data = country_dict['China']
zim_data = country_dict['Zimbabwe']
ger_data = country_dict['Germany']
mex_data = country_dict['Mexico']
usa_data = country_dict['United States of America']

print(chile_data)

#Visualization
plt.scatter(chile_data.life_expectancy, chile_data.gdp)
plt.scatter(china_data.life_expectancy, china_data.gdp)
plt.scatter(zim_data.life_expectancy, zim_data.gdp)
plt.scatter(ger_data.life_expectancy, ger_data.gdp)
plt.scatter(mex_data.life_expectancy, mex_data.gdp)
plt.scatter(usa_data.life_expectancy, usa_data.gdp)
plt.legend(['Chile','China','Zimbabwe','Germany','Mexico', 'USA'])
plt.xlabel('Life Expetancy at Birth in Years')
plt.ylabel('GDP in Trillion Dollars')
plt.title('Life Expectancy VS GDP in all 6 countries')
plt.savefig('leVgdp.png')
plt.show()
plt.clf()

#From the scatterplot, it looks like USA has the highest GDP, while Highest life expectancy is in Germany despite having less GDP compared to the USA

```

```python
#Average life expectancy and GDP in the same figure of all countries all visualized

def convert_trillion(number):
    return number * 1000000000000
le_list = []
gdp_list = []
country_names_list = ['Chile','China','Germany','Mexico','USA','Zimbabwe']
list_ticks = [1,2,3,4,5,6]

for country in list_country:
    le_country = data[data['country'] == country].life_expectancy
    gdp_country = data[data['country'] == country].gdp
    avg_le = np.mean(le_country)
    avg_gdp = np.mean(gdp_country)
    le_list.append(avg_le)
    gdp_list.append(avg_gdp)

ax = plt.subplot()


plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)
plt.subplot(1,2,1)
plt.bar(country_names_list, le_list, color='r')
plt.xticks(rotation = 35)
plt.xlabel('Country')
plt.ylabel('Age in Years')
plt.ylim(45,85)
plt.title('Average LE in Countries')

plt.subplot(1,2,2)
plt.bar(country_names_list,gdp_list, color='g')
plt.xticks(rotation = 35)
plt.xlabel('Country')
plt.ylabel('Gross Domestic Product in Trillion $')
plt.title('Average GDP in Countries ')
plt.savefig('LE_GDP.png')
plt.show()
plt.clf()
    
    
# Highest average life expectancy is Germany 
#Lowest average Life expectancy is Zimbabwe

```

```python
#scatter plot of all the countries


plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)


plt.subplot(2,3,1)
plt.scatter(chile_data.life_expectancy, chile_data.gdp,color='r')
plt.title('Chile')

plt.subplot(2,3,2)
plt.scatter(china_data.life_expectancy, china_data.gdp, color='blue')
plt.title('China')

plt.subplot(2,3,3)
plt.scatter(zim_data.life_expectancy, zim_data.gdp, color='black')
plt.title('Zimbabwe')

plt.subplot(2,3,4)
plt.scatter(ger_data.life_expectancy, ger_data.gdp,color='green')
plt.title('Germany')

plt.subplot(2,3,5)
plt.scatter(mex_data.life_expectancy, mex_data.gdp, color='purple')
plt.title('Mexico')

plt.subplot(2,3,6)
plt.scatter(usa_data.life_expectancy, usa_data.gdp)
plt.title('USA')
plt.savefig('separete_scatter.png')

plt.show()
plt.clf()
#Find correlation

def find_correlation(data):
    le_list = list(data['life_expectancy'])
    gdp_list = list(data['gdp'])
    r = np.corrcoef(le_list,gdp_list)
    return(r[0,1])

data_list = [chile_data, china_data, zim_data, ger_data, mex_data, usa_data ]
correlation_dict = {}
country_n_list = []
country_r_list =[]

for data in data_list:
    country_name_list = data.country.unique()
    country_from_list = country_name_list[0]
    country_n_list.append(country_from_list)
    country_r_list.append(np.round(find_correlation(data),2))
    
correlation_dict = {'Country':country_n_list, 'Correlation':country_r_list }
correlation_df = pd.DataFrame.from_dict(correlation_dict)

plt.bar(correlation_df.Country, correlation_df.Correlation, color='g')
plt.xticks(rotation=30)
plt.ylim(0.90, 1.00)
plt.title('Correlation between Life Expectancy and GDP ')
plt.ylabel('Correlation Coefficient')
plt.savefig('correlation_var.png')
plt.show()
plt.clf()
      






```
